```python
#Divide Independent and Dependent Features
X = df.drop(labels=['default payment next month'],axis=1)
y = df['default payment next month']
```

```python
X.head()
```

```python
y
```

```python
#Train test split 
from sklearn.model_selection import train_test_split

XX_train , XX_test , yy_train , yy_test = train_test_split(X,y,test_size=30,random_state=42)
```

```python
XX_train
```

```python
#Scaling the Data 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #Instance 
```

```python
X_train = scaler.fit_transform(XX_train)
X_test = scaler.transform(XX_test)
```

```python
X_train
```

```python
X_test
```

```python
#Make data frame of scaled data
scaled_train_data = pd.DataFrame(X_train,columns=XX_train.columns,index=XX_train.index)
scaled_test_data = pd.DataFrame(X_test,columns=XX_test.columns,index=XX_test.index)
```

```python
scaled_train_data.head()
```

```python
scaled_test_data.head()
```

```python
## Model Training 
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

```python
#Creating Instance 

nb = GaussianNB()
rf = RandomForestClassifier()
```

```python
#Now fit the data 
nb.fit(scaled_train_data,yy_train)
rf.fit(scaled_train_data,yy_train)
```

```python
#Import metrics 
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

#Prediction For Naive Bayes 
y_pred_nb = nb.predict(scaled_test_data)

#Performance For Naive Bayes
print(f"Confusion Matrix Naive bayes :  {confusion_matrix(yy_test,y_pred_nb)}")
print(f"Classification Report  Naive bayes:{classification_report(yy_test,y_pred_nb)}")
print(f"Accuracy Score Naive bayes :  {accuracy_score(yy_test,y_pred_nb)}")
```

```python
#Prediction For Random Forest
y_pred_nb = rf.predict(scaled_test_data)

#Performance For Random Forest 
print(f"Confusion Matrix Random Forest:  {confusion_matrix(yy_test,y_pred_nb)}")
print(f"Classification Report  Random Forest:  {classification_report(yy_test,y_pred_nb)}")
print(f"Accuracy Score Random Forest :  {accuracy_score(yy_test,y_pred_nb)}")
```

```python
#Hyperparameter Tunning 
from sklearn.model_selection import GridSearchCV
param_grid = {'var_smoothing':[0.1,0.01,0.5,0.005,0.001,1e-6,1e-8,1e-10,1e-11]}
grid = GridSearchCV(estimator=nb,param_grid=param_grid,cv=5,verbose=3)
```

```python
grid.fit(scaled_train_data,yy_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
model_nb = GaussianNB(var_smoothing=0.5)
model_nb.fit(scaled_train_data,yy_train)
y_pred_nb = model_nb.predict(scaled_test_data)

#Performance For Naive Bayes
print(f"Confusion Matrix Naive bayes :  {confusion_matrix(yy_test,y_pred_nb)}")
print(f"Classification Report  Naive bayes:{classification_report(yy_test,y_pred_nb)}")
print(f"Accuracy Score Naive bayes :  {accuracy_score(yy_test,y_pred_nb)}")
```
